```python
from pymatgen.analysis.structure_matcher import StructureMatcher, ElementComparator
```

```python
        def oqmd2pymatgen_struct(d):
            sites = []
            lat = Lattice(d['unit_cell'])
            for s in d['sites']:
                sp, _, x, y, z = s.split()
                x, y, z = map(float, [x, y, z])
                site = PeriodicSite(sp, [x, y, z], lat)
                sites.append(site)
            struct = Structure.from_sites(sites)
            return struct
```

```python
S_cutoff = 0.3
sm = StructureMatcher(comparator=ElementComparator(), attempt_supercell=True)
for i in range(len(ordered_entries)):
    s1 = ordered_entries[i].structure
    if len(s_total[i]['data']) !=0:
        for j in range(len(s_total[i]['data'])):
            s2 = oqmd2pymatgen_struct(s_total[i]['data'][j])
            # groups = sm.group_structures([s1, s2])
            if sm.fit(s1, s2):
                ordered_entries[i].stability = s_total[i]['data'][j]['stability']
                break

no_stability = [_ for _ in ordered_entries if _.stability is None]  # should be removed at last
stable_entires = [_ for _ in ordered_entries if _.stability is None or _.stability < S_cutoff]
all_entires = [_ for _ in precess.entries if _.stability is None or _.stability < S_cutoff]
```

```python
    def get_dataframe(icdd_entries, keys):
        data = {}
        for key in keys:
            data[key] = [e.as_dict()[key] for e in icdd_entries]
        df = pd.DataFrame(data)
        return df
```

```python
df = get_dataframe([_ for _ in stable_entires],
                   ['entry_id', 'name', 'pressure_temperature', 'cross_refs', 'status', 'quality_mark', 'name',
                    'spgr', 'common_name', 'leader',  'stability'])
df_all = get_dataframe([_ for _ in all_entries],
                   ['entry_id', 'name', 'pressure_temperature', 'cross_refs', 'status', 'quality_mark', 'name',
                    'spgr', 'common_name', 'leader',  'stability'])
print(df)
print('[ICDD] entries without stability', [_.entry_id for _ in no_stability])
print('[ICDD] entries with stability', df_all)
df.to_excel('./data/output_candidate_pool.xlsx')

with open('./data/entries_dft.json', 'w') as f:
        json.dump(all_entries, f, cls=MontyEncoder)
```
