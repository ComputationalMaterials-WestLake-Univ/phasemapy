```python
def q_to_d (q):
    import numpy as np
    d = 2*np.pi/q
    return d

def simulated_dps(d,entry_id,intensity_log,cutoff=7):
    dps_id = []
    dps = np.round(1000/np.array(d[np.where(intensity_log>=cutoff)])-50)
    for i in range(len(dps)): 
        dps_id.append(tuple([dps[i], entry_id]))    
    return dps_id

def exprimental_dps(d,intensity,cutoff=2,plot=True):
    import scipy.signal
    import matplotlib.pyplot as plt
    x = intensity/max(intensity)*100
    peaks, _ = scipy.signal.find_peaks(x,height=cutoff,)
    exp_d = d[peaks]
    exp_i = x[peaks]
    exp_dps = np.round(1000/np.array(exp_d)-50) 
    if plot:
        plt.plot(d,x)
        plt.plot(d[peaks], x[peaks], "o",)
        plt.title("scipy.signal.find_peaks")
        plt.show()
    return exp_dps,exp_d,exp_i

def sort_dps(dps):
    from operator import itemgetter
    sorted_dps = sorted(dps, key=itemgetter(0))
    return sorted_dps

def search_dps(simu_dps_intensity, exp_dps,window=30):
    phases = []
    simu_dps = []
    for i in range(len(simu_dps_intensity)):
        simu_dps.append (simu_dps_intensity[i][0])
    simu_dps = np.array(simu_dps)
    for i in range(len(exp_dps)):
        idx_within_window = np.where(np.abs(simu_dps - exp_dps[i]) <= window)
        if len(idx_within_window) !=0:
            closest_values = simu_dps[idx_within_window]
            phases_id = np.array(sorted_dps)[idx_within_window][:,1]
            phases_id = phases_id.tolist()
            phases.append(phases_id)
    from collections import Counter
    phases_arr = [item for p in phases for item in p if item]
    phases_count = Counter(phases_arr)
    return phases_count

def linear_to_log(intensity):
    intensity[intensity < 1] = 1
#     intensity[(intensity <= 10) & (intensity > 7) ] = 7
    intensity[intensity >64] = 64
    
    return np.round(5*np.log10(intensity))
```
