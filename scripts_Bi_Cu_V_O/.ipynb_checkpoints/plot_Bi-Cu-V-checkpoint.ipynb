```python
activated_entries =set()
for sample in samples:
    sample_entries = [_ for _ in sample.entries if _.name != 'SnO2']
    activated_entries = activated_entries | set(sample_entries)

max_act = defaultdict(float)
tot_act = defaultdict(float)
count_act = defaultdict(int)
ref = defaultdict(list)
for sample in samples:
    norm = sum([sample.height_fractions[i] for i,p in enumerate(sample.solution) if p.entry.name != 'SnO2'])
    for index,phase in enumerate(sample.solution):
        tot_act[phase.entry.entry_id]+=sample.height_fractions[index]/norm
        count_act[phase.entry.entry_id]+=1
        max_act[phase.entry.entry_id] = max(sample.height_fractions[index]/norm,max_act[phase.entry.entry_id])
        ref[phase.entry.entry_id].append([sample.height_fractions[index]/norm,sample.sample_id])
        ref[phase.entry.entry_id] = sorted(ref[phase.entry.entry_id],reverse=True)
```

```python
def get_phase_fraction(instance_data, sequence, ncols, nrows, samples,entry_ids):
    chemsys = instance_data.chemsys
    i, j = chemsys.index(sequence[-1]), chemsys.index(sequence[0])
    unitvec = np.array([[1, 0], [0.5, np.sqrt(3) / 2]])
    x, y = np.dot(instance_data.sample_comp[:, (i, j)], unitvec).T
    vmin=1-0.02
    vmax = 1+0.02
    cm = 1/2.54

    fig, ax = plt.subplots(1, 1, figsize=(2.9*3*cm,2.9*3*cm),)
        
    phase_all=[]    
    for i,  entry_id in enumerate(entry_ids): 
        entry=entry_dict[entry_id]       
        ax.scatter(x, y, s=5, c='white', alpha=0.4, lw=0)
        label = label_list[i]
        phase_amt = []
        shifts = []
        for sample in samples:
            if entry in sample.entries:
                index = sample.entries.index(entry)
                fraction = sample.height_fractions[index]
                shift = sample.solution[index].shift+1
                if abs(sample.solution[index].shift)>0.015:
                    shift = 1 
                if sample.height_fractions[index]<0.01:
                    fraction=0
                phase_amt.append(fraction)
                shifts.append(shift)
            else:
                phase_amt.append(0)
                shifts.append(1)
                    


        phase_all.append(phase_amt)
    phase_all = np.array(phase_all)
    print(phase_all)
    return phase_all

   
    
all_entries = list(activated_entries)        
entry_dict = {e.entry_id:e for e in all_entries}
sequence = ['Bi', 'Cu', 'V']  # up, left, right

phase_all = get_phase_fraction(instance_data, sequence, 3, 5, samples,list(df['entry_id']))

# plt.savefig('phase_diagram_Bi-Cu-V.pdf',format='pdf',bbox_inches='tight',transparent=True)
# plt.savefig('phase_diagram_Bi-Cu-V.svg',format='svg',bbox_inches='tight',transparent=True)
```
